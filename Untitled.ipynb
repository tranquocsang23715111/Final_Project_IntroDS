```python
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer

# hàm đếm âm tiết
def dem_am_tiet(tu):
    tu = tu.lower()
    am_tiet = re.findall(r'[aeiouy]+', tu)
    return max(1, len(am_tiet))

# điểm dễ đọc
def diem_flesch(van_ban):
    cau = re.split(r'[.!?]', str(van_ban))
    cau = [c for c in cau if c.strip()]
    so_cau = len(cau) or 1

    tu = re.findall(r'\w+', str(van_ban))
    so_tu = len(tu) or 1

    tong_am_tiet = sum(dem_am_tiet(t) for t in tu)

    diem = 206.835 - 1.015 * (so_tu / so_cau) - 84.6 * (tong_am_tiet / so_tu)
    return diem

# chấm mức độ dê hiểu
def diem_ro_rang(van_ban):
    diem = diem_flesch(van_ban)
    if diem > 80:
        return 5
    elif diem > 70:
        return 4
    elif diem > 60:
        return 3
    elif diem > 50:
        return 2
    else:
        return 1

# chấm mức độ lên quan
def diem_lien_quan(cau_hoi, cau_tra_loi):
    vectorizer = CountVectorizer(stop_words='english')
    tu_hoi = vectorizer.build_tokenizer()(str(cau_hoi).lower())
    tu_tra_loi = vectorizer.build_tokenizer()(str(cau_tra_loi).lower())

    if not tu_hoi:
        return 1

    tu_giong_nhau = [tu for tu in tu_hoi if tu in tu_tra_loi]
    ti_le_trung = len(tu_giong_nhau) / len(set(tu_hoi)) * 100

    if ti_le_trung >= 0.80:
        return 5
    elif ti_le_trung >= 0.60:
        return 4
    elif ti_le_trung >= 0.40:
        return 3
    elif ti_le_trung >= 0.20:
        return 2
    else:
        return 1

# chấm độ đầy đủ

def diem_day_du(cau_tra_loi):
    so_tu = len(str(cau_tra_loi).split())
    if so_tu > 40:
        return 5
    elif so_tu > 20:
        return 4
    elif so_tu > 10:
        return 3
    elif so_tu > 4:
        return 2
    else:
        return 1

#đọc dl
duong_dan_file = 'DChatGPT.xlsx'  # đổi tên nếu cần
du_lieu_goc = pd.read_excel(duong_dan_file, sheet_name='Sheet1')

# Loại bỏ dòng đầu tiên
du_lieu = du_lieu_goc.iloc[1:].copy()
du_lieu.columns = du_lieu_goc.iloc[0]

#chấm điểm
du_lieu['RoRang'] = du_lieu['Answer'].apply(diem_ro_rang)
du_lieu['LienQuan'] = du_lieu.apply(lambda dong: diem_lien_quan(dong['Generated Question'], dong['Answer']), axis=1)
du_lieu['DayDu'] = du_lieu['Answer'].apply(diem_day_du)

#điểm tb
du_lieu['TongDiem (1-5)'] = du_lieu[['RoRang', 'LienQuan', 'DayDu']].mean(axis=1).round(2)

#lưu vào file
du_lieu.to_excel('DChatGPT_cham_diem.xlsx', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cấu hình hiển thị
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
df = pd.read_excel('DChatGPT_cham_diem.xlsx')
score_cols = ['RoRang', 'LienQuan', 'DayDu', 'TongDiem (1-5)']

for col in score_cols:
    plt.figure()
    sns.histplot(df[col], bins=5, kde=True, color="skyblue")
    plt.title(f'Phân bố điểm {col}')
    plt.xlabel('Điểm')
    plt.ylabel('Số lượng')
    plt.xticks([1, 2, 3, 4, 5])
    plt.grid(True)
    plt.show()
avg_scores = df[score_cols].mean().sort_values()


plt.figure()
sns.heatmap(df[score_cols].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Ma trận tương quan giữa các tiêu chí đánh giá')
plt.show()

```
